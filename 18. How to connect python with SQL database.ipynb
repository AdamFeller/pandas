# 18. How to connect pythob with SQL DataBase

Source: https://stackoverflow.com/questions/33725862/connecting-to-microsoft-sql-server-using-python

```python
import pyodbc  
import pandas as pd

COM_cnxn = pyodbc.connect("Driver={SQL Server Native Client 11.0};"
                      "Server=BDPRADA.LACAISSE.COM;"
                      "Database=ADA_CSTAT;"
                      "Trusted_Connection=yes;")

MKT_cnxn = pyodbc.connect("Driver={SQL Server Native Client 11.0};"
                      "Server=BDPRADA.LACAISSE.COM;"
                      "Database=ADA_MB;"
                      "Trusted_Connection=yes;")



company = pd.read_sql_query('select * from dbo.company', COM_cnxn)

MKT = pd.read_sql_query('select * from [cstat_base].[market_cap_monthly]', MKT_cnxn)

```

### Concatenate both tables

```python
company.set_index('gvkey', inplace = True)
MKT.set_index(['gvkey','datadate'], inplace = True)
```

```python
company_name = pd.concat([company,MKT], axis = 1)
```
