```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies.dtypes
```

```python
movies.genre.describe()
```

```python
movies.groupby('genre').genre.count().sort_values(ascending = False)
```

```python
movies.genre.value_counts()
```

```python
movies.genre.value_counts(normalize =True)
```

```python
movies.genre.unique()
```

```python
movies.genre.nunique()
```

```python
pd.crosstab(movies.genre, movies.content_rating)
```

```python
pd.crosstab(movies.genre, movies.content_rating, normalize =True)
```

```python
movies.duration.describe()
```

```python
%matplotlib inline
```

```python
movies.duration.value_counts().head()
```

```python
movies.duration.value_counts().sort_index(axis = 0).plot(kind='bar')
```

```python
movies.duration.plot(kind='hist')
```
